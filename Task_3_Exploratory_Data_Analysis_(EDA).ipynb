```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
netflix = pd.read_csv("/content/netflix_titles.csv")
iris = pd.read_csv("/content/Iris.csv")
```

```python
print(netflix.head())
print(netflix.info())

print(iris.head())
print(iris.info())
```

```python
iris.drop(columns=["Id"]).hist(figsize=(10,6))
plt.suptitle("Distribution of Iris Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris.drop(columns=["Id"]))
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
plt.figure(figsize=(8,6))

numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_iris.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="type", data=netflix)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="rating", data=netflix, order=netflix["rating"].value_counts().index)
plt.title("Netflix Content Ratings")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(netflix["release_year"], bins=30, kde=True)
plt.title("Content Release Year Distribution")
plt.show()
```

```python
netflix_movies = netflix[netflix["type"] == "Movie"]
netflix_movies["duration_num"] = netflix_movies["duration"].str.replace(" min", "").astype(float)

plt.figure(figsize=(8,5))
sns.boxplot(x=netflix_movies["duration_num"])
plt.title("Movie Duration Outliers")
plt.show()
```
